Handling PDF Challenges

Purpose of this section PDFs are notorously difficult to parse because they:
1. Store text in complex way(not just simple text)
2. Can have formatting issues
3. May contain scanned images instead of text(recuiring OCR which means we need to use a library that supports OCR)
4. Often have extracttion artifacts like headers, footers, page numbers, etc. that can interfere with text extraction and chunking.

```python
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from typing import List
```

```python
class SmartPDFProcessor:
    """Advanced PDF processing class that intelligently selects the best loader and applies cleaning and splitting strategies."""
    def __init__(self, chunk_size=1000, chunk_overlap=200):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=self.chunk_size, 
            chunk_overlap=self.chunk_overlap,
            separators=[" "]
        )

    def process_pdf(self, pdf_path:str)->List[Document]:
        """Process pdf smart chunking and metadata enhancement."""

        # Load pdf

        loader = PyPDFLoader(pdf_path)
        pages = loader.load()

        # Process each page

        processed_chunks = []
        for page_num, page in enumerate(pages):
            # Clean text
            cleaned_text = self._clean_text(page.page_content)

            if len(cleaned_text.strip()) < 50:
                continue  # Skip pages with very little content

            # Create chunks with enhanced metadata
            chunks = self.text_splitter.create_documents(
                texts = [cleaned_text], 
                metadatas=[{
                    **page.metadata,
                    "page": page_num + 1,  # Page numbers are 1-indexed
                    "total_pages": len(pages),
                    "chunk_method": "smart_pdf_processor",
                    "char_count": len(cleaned_text)
                }]
            )
            processed_chunks.extend(chunks)

        return processed_chunks

    def _clean_text(self, text:str)->str:
        """Clean text by removing excessive whitespace and common artifacts."""
        cleaned_text = ' '.join(text.split())
        return cleaned_text
    
```

```python
preprocessor = SmartPDFProcessor()
```

```python
# Process a pdf if available

try:
    smart_chunks = preprocessor.process_pdf("data/pdf/attention.pdf")
    print(f"Processed {len(smart_chunks)} smart chunks from the PDF.")

    if smart_chunks:
        print("\nSample chunk metadata:")
        for i, chunk in smart_chunks[0].metadata.items():
            print(f"{i}: {chunk}")
except Exception as e:
    print(f"Error processing PDF: {e}")
```
