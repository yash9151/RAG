Directory Loader - Reading all text files from Directory

```python
### I want to read all text files in the directory. How can I do that?

### Directory Loader
from langchain_community.document_loaders import DirectoryLoader

## load all the text files from the directory
dir_loader=DirectoryLoader(
    "data/text_files",
    glob="**/*.txt", ## Pattern to match files  
    loader_cls= TextLoader, ##loader class to use
    loader_kwargs={'encoding': 'utf-8'},
    show_progress=False
)

documents=dir_loader.load()
print("Loaded Documents:", len(documents))
documents
```

```python
for i,doc in enumerate(documents):
    print(f"Document {i+1} Content:", doc.page_content[:200])  # Print first 200 characters
    print(f"metadata: {doc.metadata}")
    print("-"*50)
```

Text Splitting Strategy

```python
""" We can split the documents into smaller chunks using different strategies. This is important for better embedding and retrieval performance. Let's explore some common text splitting strategies. 
1. Character-based splitting - it will split the text based on characters and it will not care about the meaning of the text. It will just split the text into chunks of a certain size.
2. Recursive character-based splitting - it will split the text recursively based on characters, ensuring that splits don't occur in the middle of sentences or paragraphs.
3. Token-based splitting - it will split the text based on tokens, which are meaningful units of text (e.g., words or subwords).
4. Custom splitting - it will allow you to define your own splitting logic based on specific requirements."""
```

```python
###Different Text Splitting Strategies
from langchain_text_splitters import (
    RecursiveCharacterTextSplitter,
    CharacterTextSplitter,
    TokenTextSplitter
)
```

```python
print(documents)
```

```python
###Method 1 - CharacterTextSplitter
text=documents[0].page_content
text
```

```python
# Method 1: Character-based splitting
print("Character-based splitting:")
character_splitter = CharacterTextSplitter(
    separator="\n",  #Split on newlines
    chunk_size=100,  # Max chunk size in characters
    chunk_overlap=20,  #Overlap between chunks in characters
    length_function=len # How to measure chunk size (using character count
)

character_chunks = character_splitter.split_text(text)
character_chunks
```

```python
print(character_chunks[0])
print("-"*50)
print(character_chunks[1])
print("-"*50)
print(character_chunks[2])
```

```python

# Method 2: Recursive Character-based splitting
print("Recursive Character-based splitting:")
recursive_character_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""],  # Try splitting on double newlines, then single newlines, then spaces, and finally characters
    chunk_size=200,
    chunk_overlap=20,
    length_function=len
)
recursive_character_chunks = recursive_character_splitter.split_text(text)
print(f"Created  {len(recursive_character_chunks)} chunks using Recursive Character-based splitting.")
print(f"First chunk:\n{recursive_character_chunks[0]}")
print(recursive_character_chunks)
```

```python
print(recursive_character_chunks[0])
print("-"*50)
print(recursive_character_chunks[1])
print("-"*50)
print(recursive_character_chunks[2])
```

```python
print(len(recursive_character_chunks))
```

```python
# Create Text without natural break points
simple_text = "this is sentence one and it is quite long. This is sentence two and it is also quite long. This is sentence three and it is also quite long."

splitter = RecursiveCharacterTextSplitter(
    separators=[" "],  # Try splitting on periods followed by space, then on spaces
    chunk_size=32,
    chunk_overlap=10,
    length_function=len
)
simple_chunks = splitter.split_text(simple_text)
print(simple_chunks)
print(f"Simple text was split into {len(simple_chunks)} chunks:")

for i in range(len(simple_chunks)-1):
    print(f"Chunk {i+1}: {simple_chunks[i]}")
    print(f"Chunk {i+2}: {simple_chunks[i+1]}")

    print()
```

```python
# Method 3: Token-based splitting - it will also focus on spaces not only
print("Token-based splitting:")
token_splitter = TokenTextSplitter(
    chunk_size=50,  # Max chunk size in tokens
    chunk_overlap=10,  # Overlap between chunks in tokens
    length_function=lambda x: len(x.split())  # Measure chunk size using token count (simple split on spaces)
)
token_chunks = token_splitter.split_text(text)
print(f"Created {len(token_chunks)} chunks using Token-based splitting.")
print(f"First chunk:\n{token_chunks[0]}")
```
