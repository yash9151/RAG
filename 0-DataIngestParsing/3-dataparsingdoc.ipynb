Word Document Processing

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
## Method 1: Using Docx2txtLoader
print("Method 1: Using Docx2txtLoader")
try:
    docx_loader = Docx2txtLoader("data/word_files/proposal.docx")
    docs = docx_loader.load()
    print(f"Number of documents loaded: {len(docs)}")
    print(f"First document content:\n{docs[0].page_content[:500]}...")  # Print first 500 characters
    print(f"Metadata of first document:\n{docs[0].metadata}")
except Exception as e:
    print(f"Error loading document with Docx2txtLoader: {e}")
```

```python
## Method 2: Using UnstructuredWordDocumentLoader
print("\nMethod 2: Using UnstructuredWordDocumentLoader")
try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/word_files/proposal.docx", mode="elements") #mode elements is better for docx files as it captures more structure
    docs_unstructured = unstructured_loader.load()
    print(f"Number of documents loaded: {len(docs_unstructured)}")
    print(f"First document content:\n{docs_unstructured[0].page_content[:500]}...")  # Print first 500 characters
    print(f"Metadata of first document:\n{docs_unstructured[0].metadata}")

    for i, doc in enumerate(docs_unstructured[:5]):  # Print first 5 documents
        print(f"\nElement {i+1}:")
        print(f"\nDocument {i+1} content:\n{doc.page_content[:500]}...")  # Print first 500 characters
        print(f"Metadata of document {i+1}:\n{doc.metadata}")
except Exception as e:
    print(f"Error loading document with UnstructuredWordDocumentLoader: {e}")
```
