```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
embeddings
print("Model loaded successfully!")
```

```python
# Single text embedding
text = "Hello, how are you?"
embedding = embeddings.embed_query(text)
print("Embedding for single text:")
print(embedding)
```

```python
# Example 2: Multiple text embeddings
texts = ["Hello, how are you?", "What is your name?", "Where are you from?"]
embeddings_list = embeddings.embed_documents(texts)
print("\nEmbeddings for multiple texts:")
for i, emb in enumerate(embeddings_list):
    print(f"Text: {texts[i]}")
    print(f"Embedding: {emb}\n")
```

```python
# Cosine Similarity using OpenAIEmbeddings
```

```python
import numpy as np
def cosine_similarity(vec1, vec2):
    """
    Cosine similarity measures the cosine of the angle between two vectors, giving a value between -1 and 1.
    A value of 1 means the vectors are identical, 0 means they are orthogonal, and -1 means they are opposite.
    """
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)
```

```python
examples = [
    "What is the capital of France?",
    "Where is the Eiffel Tower located?",
    "What is the largest mammal?"
]
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
embeddings
```

```python
sentence_embeddings = embeddings.embed_documents(examples)
sentence_embeddings
```

```python
# Calculate cosine similarity between the first and second example
similarity = cosine_similarity(sentence_embeddings[0], sentence_embeddings[1])
print(f"Cosine similarity between example 1 and example 2: {similarity}")
```
